```python
rank = []
state = []
gsdp_1920 = []
gsdp_1819 = []
share = []
gdp = []

for i in driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[1]'):
    if i.text == '':
        rank.append('-')
    else:
        rank.append(i.text)
for i in driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[2]'):
    if i.text == '':
        state.append('-')
    else:
        state.append(i.text)
for i in driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[3]'):
    if i.text == '':
        gsdp_1920.append('-')
    else:
        gsdp_1920.append(i.text)        
for i in driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[4]'):
    if i.text == '':
        gsdp_1819.append('-')
    else:
        gsdp_1819.append(i.text)
for i in driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[5]'):
    if i.text == '':
        share.append('-')
    else:
        share.append(i.text)
for i in driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[6]'):
    if i.text == '':
        gdp.append('-')
    else:
        gdp.append(i.text)
        
        
        
print(len(rank),len(state),len(gsdp_1920),len(gsdp_1819),len(share),len(gdp))
```

# Q5 - Gitbhub - Completed

```python
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome(r'C:\Web Driver\chromedriver.exe')
time.sleep(5)
driver.get('https://github.com/')

#hover to explore menu
action = ActionChains(driver)
menu = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]')
action.move_to_element(menu).perform()

# click trending
click_button = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')
click_button.click()
```

```python
repo = []
desc = []
count = []
lang = []

for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
    repo.append(i.text)
for i in driver.find_elements_by_xpath('//p[@class="col-9 color-fg-muted my-1 pr-4"]'):
    desc.append(i.text)
for i in driver.find_elements_by_xpath('//a[@class="Link--muted d-inline-block mr-3"][2]'):
    count.append(i.text)
for i in driver.find_elements_by_xpath('//div[@class="f6 color-fg-muted mt-2"]/span[1]'):
    lang.append(i.text)

df = pd.DataFrame({'Repository':repo,'Description':desc,'Contributors':count,'Language Used':lang})
df['Language Used'] = df['Language Used'].replace(to_replace='Built by', value='-')
df
```
