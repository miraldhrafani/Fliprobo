```python
from selenium.common.exceptions import InvalidSelectorException
```

```python
driver = webdriver.Chrome(r'C:\Web Driver\chromedriver.exe')
time.sleep(2)
driver.get('https://www.flipkart.com/')

# Closing the login Page
close_button = driver.find_element_by_xpath("/html/body/div[2]/div/div/button")
close_button.click()

# enetering 'Sunglasses' in search product column
product_search = driver.find_element_by_class_name('_3704LK')
product_search.send_keys('smartphone')

# clicking on search button
search_button = driver.find_element_by_class_name('L0Z3Pu')
search_button.click()
```

```python
# creating empty list for data to be entered
brand = []
name = []
colour = []
ram = []
storage = []
pcamera = []
scamera = []
display = []
battery = []
price = []
url = []

#searching for brand and appending into the list
for i in driver.find_elements_by_xpath('//div[@class="_4rR01T"]'):
    brand.append(''.join(i.text.split()[0]))
    
#searching for name and appending it into list
for i in driver.find_elements_by_xpath('//div[@class="_4rR01T"]'):
    name.append(i.text)

#searching for colour and appending it into list
for i in driver.find_elements_by_xpath('//div[@class="_4rR01T"]'):
    colour.append(' '.join(i.text.split()[-4:-2])[1:-1])
    
#searching for ram and appending it into list
for i in driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]/li[1]'):
    ram.append( ' '.join(i.text.split()[:3]))
    
#searching for storage and appending it into list
for i in driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]/li[1]'):
    storage.append( ' '.join(i.text.split()[-8:-5]))
    
#searching for primary camera and appending it into list
for i in driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]/li[3]'):
    pcamera.append(i.text)
    
#searching for display camera and appending it into list
for i in driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]/li[2]'):
    display.append(i.text)
    
#searching for battery camera and appending it into list
for i in driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]/li[4]'):
    battery.append(i.text)
    
#searching for price camera and appending it into list
for i in driver.find_elements_by_xpath('//div[@class="_30jeq3 _1_WHN1"]'):
    price.append(i.text)
    
#searching for price camera and appending it into list
for i in driver.find_elements_by_xpath('//a[@class="_1fQZEK"]'):
    url.append(i.get_attribute('href'))

    
#searching for secondary camera and appending it into list
y = 25
try :
    for i in driver.find_elements_by_id('//ul[@class="_1xgFaf"]/li[10]'):
        scamera.append(i.text)    
except InvalidSelectorException:
    for i in range(1,25):
        scamera.append('-')

df = pd.DataFrame({'Brand':brand,'Phone':name,'RAM':ram,'Storage':storage,'Primary Camera':pcamera,
                  'Secondary Camera':scamera,'Display':display,'Battery':battery,'Price':price,
                  'URL':url})
df
```
