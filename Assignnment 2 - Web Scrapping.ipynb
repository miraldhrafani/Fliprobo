# Q2) IMDB top rated 100 movies

```python
page = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
content = BeautifulSoup(page.content)
```

```python
movies = []
for i in content.find_all('td',class_="titleColumn"):
    y = ' '.join(i.text.split()[1:-1])
    movies.append(y)

movies = movies[0:100]

year = []
for i in content.find_all('td',class_="titleColumn"):
    y = i.text.split()[-1]
    year.append(y.strip('()'))
    
year = year[0:100]

rating = []
for i in content.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.strip('\n'))
    
rating = rating[0:100]

df = pd.DataFrame({'Movies':movies,'Rating':rating,'Year of Release':year})
df
```

# Q5) Cricket Ranking - Men

## a)- Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
content = BeautifulSoup(page.content)
```

```python
# inserting team name
team = []

for i in content.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)

# fetching top 10 teams
team = team[:10]
## print('-- Top ten Teams --\n',team)

# inserting matches of team No 2 till team No 9
matches = []
for i in content.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text)

# fetching matches of top 10 teams and separting points from it
matches = matches[:18:2]

# inserting matches of Team 1 because it was located in different class
first = content.find('td',class_="rankings-block__banner--matches")
matches.insert(0,first.text)
## print('-- Matches --\n',matches)

# inserting points of team No 2 till team No 9
points = []
for i in content.find_all('td',class_="table-body__cell u-center-text"):
    points.append(i.text)

# fetching points of top 10 teams and separting matches from it
points = points[1:18:2]

# inserting points of Team 1 because it was located in different class
first = content.find('td',class_="rankings-block__banner--points")
points.insert(0,first.text)
## print('-- points --\n',points)

# inserting ratings of team No 2 till team No 9
rating = []
for i in content.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
    
# fetching ratings of top 10 teams
rating = rating[0:9]

first = content.find('td',class_="rankings-block__banner--rating u-text-right")
rating.insert(0,first.text.split()[0])
## print('-- rating --\n',rating)

top_ten_team = pd.DataFrame({'Teams':team,'Matches_played':matches,'Points':points,'Rating':rating})

top_ten_team['Ranking'] = np.arange(1,11,1)
top_ten_team
```

# Q7) News Headlines from Cnbc News

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
content = BeautifulSoup(page.content)
```

```python
headlines = []
for i in content.find_all('a',class_="LatestNews-headline"):
    headlines.append(i.text)

time = []
for i in content.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)

url = []
for i in content.find_all('a',class_="LatestNews-headline"):
    url.append(i['href'])
    
df = pd.DataFrame({'Headlines':headlines,"Time":time,"Link":url})
df
```

# Q9) Restaurant List of Dineout.co.in

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
content = BeautifulSoup(page.content)
```

```python
restaurant = []
for i in content.find_all('a',class_="restnt-name ellipsis"):
    restaurant.append(i.text)

cuisine = []
for i in content.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('| ',1)[1])

Location = []
for i in content.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
    
rating = []
for i in content.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
    
image_url = []
for i in content.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
    
df = pd.DataFrame({"Restaurant":restaurant,"Cuisine":cuisine,"Location":Location,
                   "Rating":rating,"Image URL":image_url})
df
```

# Q10) top publication

```python
page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
content = BeautifulSoup(page.content)
```

```python
rank = []
for i in content.find_all('td',class_="gsc_mvt_p"):
    rank.append(i.text)
    
publication = []
for i in content.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)
    
h5index = []
for i in content.find_all('td',class_="gsc_mvt_n"):
    h5index.append(i.text)
    
h5index = h5index[::2]

h5median = []
for i in content.find_all('td',class_="gsc_mvt_n"):
    h5median.append(i.text)
h5median = h5median[1::2]

top_publication = pd.DataFrame({'Rank':rank,'Publication':publication,
                                        'h-5 Index': h5index, 'H-5 Median':h5median})
top_publication
```
