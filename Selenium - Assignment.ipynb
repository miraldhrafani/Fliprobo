# Q10 Ambition Box - Salary - Completed

```python
# Opening the driver
driver = webdriver.Chrome(r'C:\Web Driver\chromedriver.exe')
time.sleep(2)


driver.get('https://www.ambitionbox.com/')
```

```python
# Clicking on Salary tabs
find = driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]')
find.click()
```

```python
# typing Data Scientist in search bar
input_key = driver.find_element_by_xpath("/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
input_key.send_keys('Data Scientist')
```

```python
# Clicking on search
find = driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')
find.click()
```

```python
#entering Data into the list
company = []
records = []
experience = []
min_sal = []
max_sal = []
avg_sal = []


for i in driver.find_elements_by_xpath('//div[@class="name"]/a'):
    company.append(i.text)
    
for i in driver.find_elements_by_xpath('//div[@class="name"]/span'):
    records.append(i.text)

for i in driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]'):
    z = ' '.join(i.text.split()[3:])
    experience.append(z)
    
for i in driver.find_elements_by_xpath('//div[@class="salary-values"]/div[1]'):
    min_sal.append(i.text)
    
for i in driver.find_elements_by_xpath('//div[@class="salary-values"]/div[2]'):
    max_sal.append(i.text)

for i in driver.find_elements_by_xpath('//p[@class="averageCtc"]'):
    avg_sal.append(i.text)
```

```python
df = pd.DataFrame({'Company':company,'Experience':experience,
                   'Min Salary':min_sal,'Max Salary':max_sal,'Average Salary':avg_sal,'Records':records})
df
```
